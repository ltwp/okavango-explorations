```python
import geopandas as gpd
import ee

def load_distributaries(shapefile_path):
    """Load distributaries shapefile as ee.FeatureCollection."""
    gdf = gpd.read_file(shapefile_path)
    features = []
    for idx, row in gdf.iterrows():
        geom = ee.Geometry.Polygon(row['geometry'].__geo_interface__['coordinates'])
        feature = ee.Feature(geom, {'dist_name': row['dist_name'], 
                                     'HYBAS_ID': row['HYBAS_ID']})
        features.append(feature)
    return ee.FeatureCollection(features)

# Usage
ee.Initialize()
distributaries = load_distributaries(r"E:\Okavango\Data\Study_Areas\Okavango_Distributaries.shp")

# Get individual distributaries
boro = distributaries.filter(ee.Filter.eq('dist_name', 'Boro'))
```

```python
import pandas as pd
import os

def get_wetted_area(distributaries, dist_name, year, month, threshold):
    """
    Calculate wetted area for a distributary in a given month.
    
    Parameters:
    -----------
    distributaries : ee.FeatureCollection
        Distributaries feature collection
    dist_name : str
        Name of distributary (e.g., 'Boro')
    year : int
        Year (e.g., 2020)
    month : int
        Month (1-12)
    threshold : int
        Minimum pixel value to include (1-4)
    
    Returns:
    --------
    dict : {'pixel_count': int, 'area_km2': float}
    """
    # Get the specific distributary
    dist = distributaries.filter(ee.Filter.eq('dist_name', dist_name))
    
    # Load the specific water mask image
    asset_path = f"projects/ee-okavango/assets/water_masks/monthly_DSWE_Landsat_30m_v4/DSWE_Products/DSWE_{year}_{month:02d}"
    water = ee.Image(asset_path).select(0).rename('water')
    
    # Threshold: include pixels >= threshold (e.g., threshold=3 gives pixels 3,4)
    water_mask = water.gte(threshold)
    
    # Calculate pixel count and area
    stats = water_mask.reduceRegion(
        reducer=ee.Reducer.sum(),
        geometry=dist.geometry(),
        scale=30,
        maxPixels=1e9
    ).getInfo()
    
    pixel_count = stats['water']
    area_km2 = pixel_count * 0.0009
    
    return {'pixel_count': pixel_count, 'area_km2': area_km2}


def calculate_monthly_inundation(distributaries, output_dir, year_start=1984, year_end=2024):
    """
    Calculate monthly inundation areas for all distributaries and save to CSV.
    
    Parameters:
    -----------
    distributaries : ee.FeatureCollection
        Distributaries feature collection
    output_dir : str
        Directory path to save CSV outputs
    year_start : int
        First year to include (default 1984)
    year_end : int
        Last year to include (default 2024)
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Get list of distributary names
    dist_names = distributaries.aggregate_array('dist_name').getInfo()
    
    # Store data for both thresholds
    data_records = []
    
    print(f"Collecting data from {year_start} to {year_end}...")
    print("This may take a while...")
    
    # Collect all available data
    total_months = (year_end - year_start + 1) * 12
    processed = 0
    
    for year in range(year_start, year_end + 1):
        for month in range(1, 13):
            processed += 1
            
            # Try to load the image (may not exist)
            try:
                # Create record for this month
                record = {'Year': year, 'Month': f'{month:02d}'}
                
                # Calculate area for each distributary for both thresholds
                for dist_name in dist_names:
                    # Threshold 1 (all inundation: pixels 1-4)
                    result_1 = get_wetted_area(distributaries, dist_name, year, month, threshold=1)
                    record[f'{dist_name}_threshold1'] = result_1['area_km2']
                    
                    # Threshold 3 (open water: pixels 3-4)
                    result_3 = get_wetted_area(distributaries, dist_name, year, month, threshold=3)
                    record[f'{dist_name}_threshold3'] = result_3['area_km2']
                
                data_records.append(record)
                print(f"  Progress: {processed}/{total_months} - {year}-{month:02d}: ✓", end='\r')
                
            except Exception as e:
                print(f"  Progress: {processed}/{total_months} - {year}-{month:02d}: missing", end='\r')
                continue
    
    print(f"\n\nCollected {len(data_records)} months of data")
    
    # Convert to DataFrame
    df_all = pd.DataFrame(data_records)
    
    # Create separate DataFrames for each threshold
    df_threshold1 = df_all[['Year', 'Month'] + [col for col in df_all.columns if 'threshold1' in col]]
    df_threshold1.columns = ['Year', 'Month'] + [col.replace('_threshold1', '') for col in df_threshold1.columns if 'threshold1' in col]
    
    df_threshold3 = df_all[['Year', 'Month'] + [col for col in df_all.columns if 'threshold3' in col]]
    df_threshold3.columns = ['Year', 'Month'] + [col.replace('_threshold3', '') for col in df_threshold3.columns if 'threshold3' in col]
    
    # Save CSVs
    print("\nSaving CSV files...")
    csv_path_all = os.path.join(output_dir, 'Distributary_InundationMonthly_AllInundation_Landsat.csv')
    csv_path_open = os.path.join(output_dir, 'Distributary_InundationMonthly_OpenWater_Landsat.csv')
    
    df_threshold1.to_csv(csv_path_all, index=False)
    df_threshold3.to_csv(csv_path_open, index=False)
    
    print(f"  ✓ Saved: {csv_path_all}")
    print(f"  ✓ Saved: {csv_path_open}")
    print(f"\n✓ Processing complete!")
    
    return df_threshold1, df_threshold3
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import os

def calculate_and_plot_zscores(csv_dir, output_dir):
    """
    Calculate monthly z-scores for each distributary based on fraction of total delta inundation.
    Outputs both figures and CSV files with z-scores.
    
    Parameters:
    -----------
    csv_dir : str
        Directory containing the input CSV files
    output_dir : str
        Directory to save output figures and CSVs
    """
    # Create output directory if needed
    os.makedirs(output_dir, exist_ok=True)
    
    # Load both CSV files
    csv_all = os.path.join(csv_dir, 'Distributary_InundationMonthly_AllInundation_Landsat.csv')
    csv_open = os.path.join(csv_dir, 'Distributary_InundationMonthly_OpenWater_Landsat.csv')
    
    df_all = pd.read_csv(csv_all)
    df_open = pd.read_csv(csv_open)
    
    print("Loaded CSV files...")
    print(f"  All Inundation: {len(df_all)} months")
    print(f"  Open Water: {len(df_open)} months")
    
    # Get distributary names (all columns except Year and Month)
    dist_names = [col for col in df_all.columns if col not in ['Year', 'Month']]
    
    # California-inspired color palette (earthy tones)
    colors = ['#E8A87C', '#C38D9E', '#41B3A3', '#7FB069', '#CD5C5C', '#8D6E63', '#5F9EA0']
    
    # Process both datasets
    datasets = [
        (df_all, 'All Inundation (≥1)', 'AllInundation'),
        (df_open, 'Open Water (≥3)', 'OpenWater')
    ]
    
    for df, title_suffix, file_suffix in datasets:
        print(f"\nProcessing {title_suffix}...")
        
        # Ensure Month is stored as zero-padded string
        df['Month'] = df['Month'].astype(str).str.zfill(2)
        df['Year'] = df['Year'].astype(int)
        
        # Create date column for plotting
        df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%m')
        
        # Calculate total delta inundation for each month-year
        df['Total_Delta'] = df[dist_names].sum(axis=1)
        
        # Calculate fraction of total for each distributary
        for dist in dist_names:
            df[f'{dist}_fraction'] = df[dist] / df['Total_Delta']
        
        # Calculate monthly statistics for fractions (mean and std for each month across all years)
        monthly_stats = {}
        for dist in dist_names:
            monthly_stats[dist] = {}
            for month in range(1, 13):
                month_str = f'{month:02d}'
                month_data = df[df['Month'] == month_str][f'{dist}_fraction'].dropna()
                monthly_stats[dist][month_str] = {
                    'mean': month_data.mean(),
                    'std': month_data.std()
                }
        
        # Calculate z-scores based on fractions
        zscore_data = {}
        for dist in dist_names:
            zscore_data[dist] = {'dates': [], 'zscores': []}
            
            for idx, row in df.iterrows():
                month = row['Month']
                fraction = row[f'{dist}_fraction']
                
                if pd.notna(fraction):
                    mean = monthly_stats[dist][month]['mean']
                    std = monthly_stats[dist][month]['std']
                    
                    # Calculate z-score (handle case where std = 0)
                    if std > 0:
                        zscore = (fraction - mean) / std
                    else:
                        zscore = 0
                    
                    zscore_data[dist]['dates'].append(row['Date'])
                    zscore_data[dist]['zscores'].append(zscore)
        
        # Create z-score DataFrame for CSV export
        print(f"  Creating z-score CSV...")
        zscore_records = []
        
        # Get all unique dates that appear in the data
        all_dates = set()
        for dist in dist_names:
            all_dates.update(zscore_data[dist]['dates'])
        
        # For each date, check if all distributaries have z-scores
        for date in sorted(all_dates):
            # Check if this date has z-scores for ALL distributaries
            has_all = True
            record = {'Year': date.year, 'Month': date.strftime('%m')}
            
            for dist in dist_names:
                if date in zscore_data[dist]['dates']:
                    idx = zscore_data[dist]['dates'].index(date)
                    record[dist] = round(zscore_data[dist]['zscores'][idx], 6)
                else:
                    has_all = False
                    break
            
            # Only add record if all distributaries have values
            if has_all:
                zscore_records.append(record)
        
        # Create DataFrame and save to CSV
        zscore_df = pd.DataFrame(zscore_records)
        zscore_csv_path = os.path.join(output_dir, f'Distributary_ZScoresMonthly_{file_suffix}_Landsat.csv')
        zscore_df.to_csv(zscore_csv_path, index=False)
        print(f"  ✓ Saved z-score CSV: {zscore_csv_path}")
        print(f"    ({len(zscore_df)} months with complete z-scores)")
        
        # Create figure with 7 subplots (4x2 grid)
        fig, axes = plt.subplots(4, 2, figsize=(6.5, 7.5))
        axes = axes.flatten()
        
        # Plot each distributary in its own panel
        for idx, dist in enumerate(dist_names):
            ax = axes[idx]
            
            # Plot z-scores
            ax.plot(zscore_data[dist]['dates'], 
                   zscore_data[dist]['zscores'],
                   color=colors[idx],
                   linewidth=1.0,
                   alpha=0.8)
            
            # Add horizontal line at z=0
            ax.axhline(y=0, color='gray', linestyle='-', linewidth=0.5, alpha=0.5)
            
            # Format x-axis
            ax.set_xlim(datetime(1984, 1, 1), datetime(2025, 12, 31))
            years = [1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2025]
            ax.set_xticks([datetime(year, 1, 1) for year in years])
            ax.set_xticklabels(years, fontname='Arial', fontsize=8)
            
            # Format y-axis
            ax.set_ylabel('Z-Score', fontname='Arial', fontsize=8)
            ax.tick_params(axis='y', labelsize=8)
            
            # Set y-axis font
            for label in ax.get_yticklabels():
                label.set_fontname('Arial')
                label.set_fontsize(8)
            
            # Title for each panel
            ax.set_title(dist, fontname='Arial', fontsize=8, fontweight='bold', pad=5)
            
            # Grid
            ax.grid(True, alpha=0.3, linewidth=0.5)
        
        # Remove extra subplot
        fig.delaxes(axes[7])
        
        # Overall title
        fig.suptitle(f'Distributary Fraction of Total Delta Inundation - {title_suffix}',
                    fontname='Arial', fontsize=8, fontweight='bold', y=0.995)
        
        # X-axis label for bottom plots
        for ax in [axes[5], axes[6]]:
            ax.set_xlabel('Year', fontname='Arial', fontsize=8)
        
        plt.tight_layout()
        
        # Save in multiple formats
        base_filename = f'Distributary_ZScores_Fraction_{file_suffix}'
        
        png_path = os.path.join(output_dir, f'{base_filename}.png')
        jpg_path = os.path.join(output_dir, f'{base_filename}.jpg')
        pdf_path = os.path.join(output_dir, f'{base_filename}.pdf')
        
        fig.savefig(png_path, dpi=1000, bbox_inches='tight')
        fig.savefig(jpg_path, dpi=1000, bbox_inches='tight')
        fig.savefig(pdf_path, format='pdf', bbox_inches='tight')
        
        print(f"  ✓ Saved: {png_path}")
        print(f"  ✓ Saved: {jpg_path}")
        print(f"  ✓ Saved: {pdf_path}")
        
        plt.show()
        plt.close(fig)
    
    print("\n✓ Z-score analysis complete!")

# Usage:
calculate_and_plot_zscores(r'E:\Okavango\Data\Inundation\Distributaries', 
                           r'E:\Okavango\Data\Inundation\Distributaries')
```

## This shows trends in z-scores for the different distributaries post-2012

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy import stats
import os

def plot_zscore_trends(csv_dir, output_dir, start_year=2013, start_month=3):
    """
    Plot z-score trends with linear regression fits from March 2013 to December 2025.
    
    Parameters:
    -----------
    csv_dir : str
        Directory containing the z-score CSV files
    output_dir : str
        Directory to save output figures
    start_year : int
        Starting year (default 2013)
    start_month : int
        Starting month (default 3 for March)
    """
    # Create output directory if needed
    os.makedirs(output_dir, exist_ok=True)
    
    # Define distributary groups and colors
    southwest = ['TKX', 'Xudum', 'Boro']
    northeast = ['Mboroga', 'Khwai', 'Magwegwe']
    panhandle = ['Xhamu']
    
    color_sw = '#FFA07A'  # Soft orange
    color_ne = '#87CEEB'  # Soft blue
    color_ph = '#000000'  # Black
    
    # Define plotting order for legend
    plot_order = ['Boro', 'TKX', 'Xudum', 'Xhamu', 'Khwai', 'Magwegwe', 'Mboroga']
    
    # Load z-score CSVs
    csv_all = os.path.join(csv_dir, 'Distributary_ZScoresMonthly_AllInundation_Landsat.csv')
    csv_open = os.path.join(csv_dir, 'Distributary_ZScoresMonthly_OpenWater_Landsat.csv')
    
    datasets = [
        (csv_all, 'All Inundation (≥1)', 'AllInundation'),
        (csv_open, 'Open Water (≥3)', 'OpenWater')
    ]
    
    for csv_path, title_suffix, file_suffix in datasets:
        print(f"\nProcessing {title_suffix}...")
        
        # Load data
        df = pd.read_csv(csv_path)
        
        # Ensure proper data types
        df['Month'] = df['Month'].astype(str).str.zfill(2)
        df['Year'] = df['Year'].astype(int)
        df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%m')
        
        # Filter to March 2013 onwards
        start_date = datetime(start_year, start_month, 1)
        df_filtered = df[df['Date'] >= start_date].copy()
        
        print(f"  Filtered to {len(df_filtered)} months from {start_date.strftime('%B %Y')} onwards")
        
        # Create figure
        fig, ax = plt.subplots(figsize=(6.5, 4))
        
        # Convert dates to numeric for regression (months since start)
        df_filtered['months_since_start'] = (df_filtered['Date'].dt.year - start_year) * 12 + \
                                            (df_filtered['Date'].dt.month - start_month)
        
        # Plot each distributary in specified order
        for dist in plot_order:
            # Check if this distributary exists in the data
            if dist not in df_filtered.columns:
                continue
            
            # Determine color
            if dist in southwest:
                color = color_sw
                label_group = 'SW'
            elif dist in northeast:
                color = color_ne
                label_group = 'NE'
            else:  # Xhamu
                color = color_ph
                label_group = 'PH'
            
            # Get data for this distributary
            dist_data = df_filtered[['Date', 'months_since_start', dist]].dropna()
            
            if len(dist_data) == 0:
                continue
            
            # Plot z-score time series
            ax.plot(dist_data['Date'], dist_data[dist], 
                   color=color, linewidth=0.8, alpha=0.4)
            
            # Calculate linear regression
            x = dist_data['months_since_start'].values
            y = dist_data[dist].values
            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            
            # Create regression line
            x_reg = dist_data['months_since_start'].values
            y_reg = slope * x_reg + intercept
            
            # Format p-value
            if p_value < 0.001:
                p_str = 'p<0.001'
            else:
                p_str = f'p={p_value:.3f}'
            
            # Create label with slope and p-value
            label = f'{dist} ({label_group}): {slope:.4f}/mo, {p_str}'
            
            # Plot regression line
            ax.plot(dist_data['Date'], y_reg, 
                   color=color, linewidth=1.5, alpha=0.9, 
                   label=label)
        
        # Add horizontal line at z=0
        ax.axhline(y=0, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
        
        # Format x-axis
        ax.set_xlim(start_date, datetime(2025, 12, 31))
        years = range(2013, 2026, 2)
        ax.set_xticks([datetime(year, 1, 1) for year in years])
        ax.set_xticklabels(years, fontname='Arial', fontsize=8)
        
        # Format y-axis
        ax.set_ylabel('Z-Score', fontname='Arial', fontsize=8)
        ax.set_xlabel('Year', fontname='Arial', fontsize=8)
        ax.tick_params(axis='both', labelsize=8)
        
        # Set tick font
        for label in ax.get_xticklabels() + ax.get_yticklabels():
            label.set_fontname('Arial')
            label.set_fontsize(8)
        
        # Title
        ax.set_title(f'Z-Score Trends by Distributary (March 2013-Dec 2025) - {title_suffix}',
                    fontname='Arial', fontsize=8, fontweight='bold', pad=10)
        
        # Grid
        ax.grid(True, alpha=0.3, linewidth=0.5)
        
        # Legend (order is preserved from plotting order)
        ax.legend(loc='best', fontsize=8, frameon=True,
                 prop={'family': 'Arial', 'size': 8})
        
        plt.tight_layout()
        
        # Save in multiple formats
        base_filename = f'Distributary_ZScore_Trends_{file_suffix}_2013-2025'
        
        png_path = os.path.join(output_dir, f'{base_filename}.png')
        jpg_path = os.path.join(output_dir, f'{base_filename}.jpg')
        pdf_path = os.path.join(output_dir, f'{base_filename}.pdf')
        
        fig.savefig(png_path, dpi=1000, bbox_inches='tight')
        fig.savefig(jpg_path, dpi=1000, bbox_inches='tight')
        fig.savefig(pdf_path, format='pdf', bbox_inches='tight')
        
        print(f"  ✓ Saved: {png_path}")
        print(f"  ✓ Saved: {jpg_path}")
        print(f"  ✓ Saved: {pdf_path}")
        
        plt.show()
        plt.close(fig)
    
    print("\n✓ Trend analysis complete!")

# Usage:
plot_zscore_trends(r'E:\Okavango\Data\Inundation\Distributaries',
                  r'E:\Okavango\Data\Inundation\Distributaries')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy import stats
import os

def plot_zscore_trends_1984_2010(csv_dir, output_dir, start_year=1984, start_month=6, 
                                 end_year=2010, end_month=5):
    """
    Plot z-score trends with linear regression fits from June 1984 to May 2010.
    
    Parameters:
    -----------
    csv_dir : str
        Directory containing the z-score CSV files
    output_dir : str
        Directory to save output figures
    start_year : int
        Starting year (default 1984)
    start_month : int
        Starting month (default 6 for June)
    end_year : int
        Ending year (default 2010)
    end_month : int
        Ending month (default 5 for May)
    """
    # Create output directory if needed
    os.makedirs(output_dir, exist_ok=True)
    
    # Define distributary groups and colors
    southwest = ['TKX', 'Xudum', 'Boro']
    northeast = ['Mboroga', 'Khwai', 'Magwegwe']
    panhandle = ['Xhamu']
    
    color_sw = '#FFA07A'  # Soft orange
    color_ne = '#87CEEB'  # Soft blue
    color_ph = '#000000'  # Black
    
    # Define plotting order for legend
    plot_order = ['Boro', 'TKX', 'Xudum', 'Xhamu', 'Khwai', 'Magwegwe', 'Mboroga']
    
    # Load z-score CSVs
    csv_all = os.path.join(csv_dir, 'Distributary_ZScoresMonthly_AllInundation_Landsat.csv')
    csv_open = os.path.join(csv_dir, 'Distributary_ZScoresMonthly_OpenWater_Landsat.csv')
    
    datasets = [
        (csv_all, 'All Inundation (≥1)', 'AllInundation'),
        (csv_open, 'Open Water (≥3)', 'OpenWater')
    ]
    
    for csv_path, title_suffix, file_suffix in datasets:
        print(f"\nProcessing {title_suffix}...")
        
        # Load data
        df = pd.read_csv(csv_path)
        
        # Ensure proper data types
        df['Month'] = df['Month'].astype(str).str.zfill(2)
        df['Year'] = df['Year'].astype(int)
        df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%m')
        
        # Filter to June 1984 through May 2010
        start_date = datetime(start_year, start_month, 1)
        end_date = datetime(end_year, end_month, 28)  # Use day 28 to capture the whole month
        df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()
        
        print(f"  Filtered to {len(df_filtered)} months from {start_date.strftime('%B %Y')} to {end_date.strftime('%B %Y')}")
        
        # Create figure
        fig, ax = plt.subplots(figsize=(6.5, 4))
        
        # Convert dates to numeric for regression (months since start)
        df_filtered['months_since_start'] = (df_filtered['Date'].dt.year - start_year) * 12 + \
                                            (df_filtered['Date'].dt.month - start_month)
        
        # Plot each distributary in specified order
        for dist in plot_order:
            # Check if this distributary exists in the data
            if dist not in df_filtered.columns:
                continue
            
            # Determine color
            if dist in southwest:
                color = color_sw
                label_group = 'SW'
            elif dist in northeast:
                color = color_ne
                label_group = 'NE'
            else:  # Xhamu
                color = color_ph
                label_group = 'PH'
            
            # Get data for this distributary
            dist_data = df_filtered[['Date', 'months_since_start', dist]].dropna()
            
            if len(dist_data) == 0:
                continue
            
            # Plot z-score time series
            ax.plot(dist_data['Date'], dist_data[dist], 
                   color=color, linewidth=0.8, alpha=0.4)
            
            # Calculate linear regression
            x = dist_data['months_since_start'].values
            y = dist_data[dist].values
            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            
            # Create regression line
            x_reg = dist_data['months_since_start'].values
            y_reg = slope * x_reg + intercept
            
            # Format p-value
            if p_value < 0.001:
                p_str = 'p<0.001'
            else:
                p_str = f'p={p_value:.3f}'
            
            # Create label with slope and p-value
            label = f'{dist} ({label_group}): {slope:.4f}/mo, {p_str}'
            
            # Plot regression line
            ax.plot(dist_data['Date'], y_reg, 
                   color=color, linewidth=1.5, alpha=0.9, 
                   label=label)
        
        # Add horizontal line at z=0
        ax.axhline(y=0, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
        
        # Format x-axis
        ax.set_xlim(start_date, datetime(end_year, 12, 31))
        years = range(1985, 2011, 5)
        ax.set_xticks([datetime(year, 1, 1) for year in years])
        ax.set_xticklabels(years, fontname='Arial', fontsize=8)
        
        # Format y-axis
        ax.set_ylabel('Z-Score', fontname='Arial', fontsize=8)
        ax.set_xlabel('Year', fontname='Arial', fontsize=8)
        ax.tick_params(axis='both', labelsize=8)
        
        # Set tick font
        for label in ax.get_xticklabels() + ax.get_yticklabels():
            label.set_fontname('Arial')
            label.set_fontsize(8)
        
        # Title
        ax.set_title(f'Z-Score Trends by Distributary (June 1984-May 2010) - {title_suffix}',
                    fontname='Arial', fontsize=8, fontweight='bold', pad=10)
        
        # Grid
        ax.grid(True, alpha=0.3, linewidth=0.5)
        
        # Legend (order is preserved from plotting order)
        ax.legend(loc='best', fontsize=8, frameon=True,
                 prop={'family': 'Arial', 'size': 8})
        
        plt.tight_layout()
        
        # Save in multiple formats
        base_filename = f'Distributary_ZScore_Trends_{file_suffix}_1984-2010'
        
        png_path = os.path.join(output_dir, f'{base_filename}.png')
        jpg_path = os.path.join(output_dir, f'{base_filename}.jpg')
        pdf_path = os.path.join(output_dir, f'{base_filename}.pdf')
        
        fig.savefig(png_path, dpi=1000, bbox_inches='tight')
        fig.savefig(jpg_path, dpi=1000, bbox_inches='tight')
        fig.savefig(pdf_path, format='pdf', bbox_inches='tight')
        
        print(f"  ✓ Saved: {png_path}")
        print(f"  ✓ Saved: {jpg_path}")
        print(f"  ✓ Saved: {pdf_path}")
        
        plt.show()
        plt.close(fig)
    
    print("\n✓ Trend analysis complete!")

# Usage:
plot_zscore_trends_1984_2010(r'E:\Okavango\Data\Inundation\Distributaries',
                             r'E:\Okavango\Data\Inundation\Distributaries')
```

## Four panel figure showing z-score trends for each distributary pre- and post-2011/12

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy import stats
import os

def plot_zscore_trends_combined_styled(csv_dir, output_dir):
    """
    Create 4-panel figure showing z-score trends with distinct line styles.
    Top row: Open Water, Bottom row: All Inundation
    Left column: 1984-2010, Right column: 2013-2025
    
    Parameters:
    -----------
    csv_dir : str
        Directory containing the z-score CSV files
    output_dir : str
        Directory to save output figures
    """
    # Create output directory if needed
    os.makedirs(output_dir, exist_ok=True)
    
    # Define distributary groups and colors
    southwest = ['TKX', 'Xudum', 'Boro']
    northeast = ['Mboroga', 'Khwai', 'Magwegwe']
    panhandle = ['Xhamu']
    
    color_sw = '#FFA07A'  # Soft orange
    color_ne = '#87CEEB'  # Soft blue
    color_ph = '#000000'  # Black
    
    # Define line styles for each distributary
    line_styles = {
        'Boro': '-',      # solid
        'TKX': '--',      # dashed
        'Xudum': ':',     # dotted
        'Xhamu': '-',     # solid
        'Mboroga': '-',   # solid
        'Khwai': '--',    # dashed
        'Magwegwe': ':'   # dotted
    }
    
    # Define plotting order
    plot_order = ['Boro', 'TKX', 'Xudum', 'Xhamu', 'Khwai', 'Magwegwe', 'Mboroga']
    
    # Load z-score CSVs
    csv_all = os.path.join(csv_dir, 'Distributary_ZScoresMonthly_AllInundation_Landsat.csv')
    csv_open = os.path.join(csv_dir, 'Distributary_ZScoresMonthly_OpenWater_Landsat.csv')
    
    # Create 2x2 subplot figure with extra space for legend
    fig, axes = plt.subplots(2, 2, figsize=(6.5, 7.5))
    
    # Define panels
    panels = [
        # (row, col, csv_path, title, period_start, period_end, panel_label)
        (0, 0, csv_open, 'Open Water (≥3)', (1984, 6), (2010, 5), 'a)'),
        (0, 1, csv_open, 'Open Water (≥3)', (2013, 3), (2025, 12), 'b)'),
        (1, 0, csv_all, 'All Inundation (≥1)', (1984, 6), (2010, 5), 'c)'),
        (1, 1, csv_all, 'All Inundation (≥1)', (2013, 3), (2025, 12), 'd)')
    ]
    
    # Store legend info (will be same for all panels)
    legend_info = []
    
    for row, col, csv_path, title, period_start, period_end, panel_label in panels:
        ax = axes[row, col]
        
        print(f"\nProcessing panel {panel_label}: {title}, {period_start[0]}-{period_end[0]}...")
        
        # Load data
        df = pd.read_csv(csv_path)
        
        # Ensure proper data types
        df['Month'] = df['Month'].astype(str).str.zfill(2)
        df['Year'] = df['Year'].astype(int)
        df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%m')
        
        # Filter to specified date range
        start_date = datetime(period_start[0], period_start[1], 1)
        end_date = datetime(period_end[0], period_end[1], 28)
        df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()
        
        print(f"  Filtered to {len(df_filtered)} months")
        
        # Convert dates to numeric for regression (months since start)
        df_filtered['months_since_start'] = (df_filtered['Date'].dt.year - period_start[0]) * 12 + \
                                            (df_filtered['Date'].dt.month - period_start[1])
        
        # Store regression stats for this panel
        panel_stats = []
        
        # Plot each distributary in specified order
        for dist in plot_order:
            # Check if this distributary exists in the data
            if dist not in df_filtered.columns:
                continue
            
            # Determine color and line style
            if dist in southwest:
                color = color_sw
                label_group = 'SW'
            elif dist in northeast:
                color = color_ne
                label_group = 'NE'
            else:  # Xhamu
                color = color_ph
                label_group = 'PH'
            
            linestyle = line_styles[dist]
            
            # Store legend info (only once)
            if row == 0 and col == 0:
                legend_info.append((dist, label_group, color, linestyle))
            
            # Get data for this distributary
            dist_data = df_filtered[['Date', 'months_since_start', dist]].dropna()
            
            if len(dist_data) == 0:
                panel_stats.append(f'{dist}: no data')
                continue
            
            # Plot z-score time series
            ax.plot(dist_data['Date'], dist_data[dist], 
                   color=color, linestyle=linestyle, linewidth=0.6, alpha=0.4)
            
            # Calculate linear regression
            x = dist_data['months_since_start'].values
            y = dist_data[dist].values
            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            
            # Create regression line
            x_reg = dist_data['months_since_start'].values
            y_reg = slope * x_reg + intercept
            
            # Plot regression line
            ax.plot(dist_data['Date'], y_reg, 
                   color=color, linestyle=linestyle, linewidth=1.2, alpha=0.9)
            
            # Format p-value
            if p_value < 0.001:
                p_str = 'p<0.001'
            else:
                p_str = f'p={p_value:.3f}'
            
            # Store stats
            panel_stats.append(f'{dist}: {slope:.4f}/mo, {p_str}')
        
        # Add horizontal line at z=0
        ax.axhline(y=0, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
        
        # Format x-axis based on period
        if period_start[0] == 1984:  # 1984-2010 period
            ax.set_xlim(start_date, datetime(2010, 12, 31))
            years = range(1985, 2011, 5)
            ax.set_xticks([datetime(year, 1, 1) for year in years])
        else:  # 2013-2025 period
            ax.set_xlim(start_date, datetime(2025, 12, 31))
            years = range(2013, 2026, 2)
            ax.set_xticks([datetime(year, 1, 1) for year in years])
        
        ax.set_xticklabels(years, fontname='Arial', fontsize=8)
        
        # Format y-axis
        ax.set_ylabel('Z-Score', fontname='Arial', fontsize=8)
        ax.tick_params(axis='both', labelsize=8)
        
        # Set tick font
        for label in ax.get_xticklabels() + ax.get_yticklabels():
            label.set_fontname('Arial')
            label.set_fontsize(8)
        
        # Panel label
        ax.text(0.02, 0.98, panel_label, transform=ax.transAxes,
               fontsize=8, fontname='Arial', fontweight='bold',
               verticalalignment='top', horizontalalignment='left')
        
        # Stats box - offset to the right of panel label
        stats_text = '\n'.join(panel_stats)
        ax.text(0.10, 0.98, stats_text, transform=ax.transAxes,
               fontsize=8, fontname='Arial',
               verticalalignment='top', horizontalalignment='left',
               bbox=dict(boxstyle='round,pad=0.3', facecolor='white', 
                        edgecolor='gray', alpha=0.8))
        
        # Title for each panel
        period_str = f"{period_start[0]}-{period_end[0]}"
        ax.set_title(f'{title} ({period_str})',
                    fontname='Arial', fontsize=8, fontweight='bold', pad=5)
        
        # Grid
        ax.grid(True, alpha=0.3, linewidth=0.5)
    
    # X-axis label for bottom plots
    for ax in [axes[1, 0], axes[1, 1]]:
        ax.set_xlabel('Year', fontname='Arial', fontsize=8)
    
    # Add legend below all panels
    # Create custom legend handles
    from matplotlib.lines import Line2D
    legend_handles = []
    legend_labels = []
    
    for dist, group, color, linestyle in legend_info:
        handle = Line2D([0], [0], color=color, linestyle=linestyle, linewidth=1.5)
        label = f'{dist} ({group})'
        legend_handles.append(handle)
        legend_labels.append(label)
    
    # Add legend below the figure
    fig.legend(legend_handles, legend_labels, 
              loc='lower center', ncol=4, fontsize=8,
              prop={'family': 'Arial', 'size': 8},
              bbox_to_anchor=(0.5, -0.02), frameon=True)
    
    plt.tight_layout(rect=[0, 0.05, 1, 1])  # Leave space for legend
    
    # Save in multiple formats
    base_filename = 'Distributary_ZScore_Trends_Combined_4Panel_Styled'
    
    png_path = os.path.join(output_dir, f'{base_filename}.png')
    jpg_path = os.path.join(output_dir, f'{base_filename}.jpg')
    pdf_path = os.path.join(output_dir, f'{base_filename}.pdf')
    
    fig.savefig(png_path, dpi=1000, bbox_inches='tight')
    fig.savefig(jpg_path, dpi=1000, bbox_inches='tight')
    fig.savefig(pdf_path, format='pdf', bbox_inches='tight')
    
    print(f"\n✓ Saved: {png_path}")
    print(f"✓ Saved: {jpg_path}")
    print(f"✓ Saved: {pdf_path}")
    
    plt.show()
    plt.close(fig)
    
    print("\n✓ Combined 4-panel styled analysis complete!")

# Usage:
plot_zscore_trends_combined_styled(r'E:\Okavango\Data\Inundation\Distributaries',
                                  r'E:\Okavango\Data\Inundation\Distributaries')
```

## Distributary map

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
from matplotlib.patches import Rectangle
from matplotlib_scalebar.scalebar import ScaleBar
import numpy as np

def create_distributary_map(shapefile_path, output_dir):
    """
    Create a map showing distributaries over satellite imagery.
    
    Parameters:
    -----------
    shapefile_path : str
        Path to distributaries shapefile
    output_dir : str
        Directory to save output map
    """
    import os
    os.makedirs(output_dir, exist_ok=True)
    
    # Load the shapefile
    gdf = gpd.read_file(shapefile_path)
    
    # Reproject to Web Mercator for basemap
    gdf_web = gdf.to_crs(epsg=3857)
    
    # California-inspired color palette (matching previous plots)
    colors = ['#E8A87C', '#C38D9E', '#41B3A3', '#7FB069', '#CD5C5C', '#8D6E63', '#5F9EA0']
    
    # Create figure (3.25 inches wide, 4 inches tall)
    fig, ax = plt.subplots(figsize=(3.25, 4))
    
    # Plot each distributary with colored fill and solid black border
    for idx, (_, row) in enumerate(gdf_web.iterrows()):
        gdf_web[gdf_web['dist_name'] == row['dist_name']].plot(
            ax=ax,
            facecolor=colors[idx],
            edgecolor='black',
            alpha=0.3,
            linewidth=2.5
        )
        
        # Add label at centroid
        centroid = row.geometry.centroid
        ax.text(centroid.x, centroid.y, row['dist_name'],
               fontsize=8, fontname='Arial', fontweight='bold',
               ha='center', va='center',
               color='white',
               bbox=dict(boxstyle='round,pad=0.3', facecolor='black', 
                        edgecolor='none', alpha=0.6))
    
    # Add satellite basemap (attribution=False removes credits)
    ctx.add_basemap(ax, source=ctx.providers.Esri.WorldImagery, zoom=11, attribution=False)
    
    # Set extent with some padding
    bounds = gdf_web.total_bounds
    padding = 0.05 * (bounds[2] - bounds[0])
    ax.set_xlim(bounds[0] - padding, bounds[2] + padding)
    ax.set_ylim(bounds[1] - padding, bounds[3] + padding)
    
    # Define specific lat/lon tick values
    lon_values = [22.0, 23.0, 24.0]
    lat_values = [-20.0, -19.0, -18.0]
    
    # Convert specific lat/lon values to Web Mercator for tick positions
    from pyproj import Transformer
    transformer = Transformer.from_crs('EPSG:4326', 'EPSG:3857', always_xy=True)
    
    # Convert to Web Mercator
    x_ticks = []
    x_labels = []
    for lon in lon_values:
        x, _ = transformer.transform(lon, -19.5)  # Use middle lat for conversion
        x_ticks.append(x)
        x_labels.append(f'{int(lon)}°E')
    
    y_ticks = []
    y_labels = []
    for lat in lat_values:
        _, y = transformer.transform(22.0, lat)  # Use middle lon for conversion
        y_ticks.append(y)
        y_labels.append(f'{int(abs(lat))}°S')
    
    # Set tick positions
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    
    # Set tick labels
    ax.set_xticklabels(x_labels, fontname='Arial', fontsize=8)
    ax.set_yticklabels(y_labels, fontname='Arial', fontsize=8)
    
    # Add scale bar with 8pt Arial
    scalebar = ScaleBar(1, location='lower right', font_properties={'family': 'Arial', 'size': 8},
                       box_alpha=0.7, color='white', box_color='black')
    ax.add_artist(scalebar)
    
    # Remove axis labels
    ax.set_xlabel('')
    ax.set_ylabel('')
    
    plt.tight_layout()
    
    # Save in multiple formats at 1000 DPI
    png_path = os.path.join(output_dir, 'Distributary_Map.png')
    jpg_path = os.path.join(output_dir, 'Distributary_Map.jpg')
    pdf_path = os.path.join(output_dir, 'Distributary_Map.pdf')
    
    fig.savefig(png_path, dpi=1000, bbox_inches='tight')
    fig.savefig(jpg_path, dpi=1000, bbox_inches='tight')
    fig.savefig(pdf_path, format='pdf', bbox_inches='tight')
    
    print(f"✓ Saved: {png_path}")
    print(f"✓ Saved: {jpg_path}")
    print(f"✓ Saved: {pdf_path}")
    
    plt.show()
    plt.close()

# Usage:
create_distributary_map(r'E:\Okavango\Data\Study_Areas\Okavango_Distributaries.shp',
                       r'E:\Okavango\Manuscript\Figures')
```
