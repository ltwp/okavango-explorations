# GEE Asset Mover

Recursively moves all assets from a source GEE folder to a destination GEE folder, preserving folder structure.

## WARNING! 
This operation will copy assets to the destination and then delete them from the source. Ensure the destination path is correct before running.

Author: James (Huck) Rees, PhD Student, UC Santa Barbara Geography

Date: December 3rd, 2024

## Import packages

```python
import ee
import time

ee.Initialize()
```

## Initialize functions

```python
def move_assets_recursively(source_folder, destination_folder, delete_source=True):
    """
    Recursively move all assets from source folder to destination folder.
    
    Args:
        source_folder (str): Full GEE source asset path (e.g., 'projects/ee-okavango/assets/water_masks/old_folder')
        destination_folder (str): Full GEE destination asset path (e.g., 'projects/ee-okavango/assets/water_masks/new_folder')
        delete_source (bool): If True, delete source assets after copying (default: True for true move operation)
    """
    
    def ensure_folder_exists(folder_path):
        """Create folder if it doesn't exist."""
        try:
            ee.data.getAsset(folder_path)
            print(f"ğŸ“ Folder exists: {folder_path}")
        except:
            # Extract parent path
            parent_path = '/'.join(folder_path.split('/')[:-1])
            folder_name = folder_path.split('/')[-1]
            
            print(f"ğŸ“ Creating folder: {folder_path}")
            ee.data.createAsset({'type': 'FOLDER'}, parent_path + '/' + folder_name)
            time.sleep(0.5)
    
    def move_recursive(source_path, dest_path):
        try:
            # List assets under the current source path
            asset_list = ee.data.listAssets({'parent': source_path})
            assets = asset_list.get('assets', [])
            
            if not assets:
                print(f"â„¹ï¸  No assets found in: {source_path}")
                return

            for asset in assets:
                source_asset_id = asset['name']
                asset_type = asset['type']
                asset_name = source_asset_id.split('/')[-1]
                
                # Construct destination path
                dest_asset_id = dest_path + '/' + asset_name

                # If it's a folder/collection, create it in destination and recurse
                if asset_type in ['FOLDER', 'IMAGE_COLLECTION', 'TABLE_COLLECTION']:
                    print(f"ğŸ“‚ Processing folder: {source_asset_id}")
                    ensure_folder_exists(dest_asset_id)
                    move_recursive(source_asset_id, dest_asset_id)
                    
                    # Delete source folder after moving its contents (if delete_source is True)
                    if delete_source:
                        try:
                            print(f"ğŸ—‘ï¸ Deleting source folder: {source_asset_id}")
                            ee.data.deleteAsset(source_asset_id)
                            time.sleep(0.5)
                        except Exception as e:
                            print(f"âš ï¸  Could not delete folder {source_asset_id}: {e}")
                else:
                    # Copy asset to destination
                    print(f"ğŸ“¦ Copying: {source_asset_id} â†’ {dest_asset_id}")
                    try:
                        ee.data.copyAsset(source_asset_id, dest_asset_id)
                        time.sleep(0.5)
                        
                        # Delete source asset after successful copy (if delete_source is True)
                        if delete_source:
                            print(f"ğŸ—‘ï¸ Deleting source: {source_asset_id}")
                            ee.data.deleteAsset(source_asset_id)
                            time.sleep(0.5)
                    except Exception as e:
                        print(f"âŒ Error copying {source_asset_id}: {e}")

        except Exception as e:
            print(f"âŒ Error processing {source_path}: {e}")

    # Ensure destination root folder exists
    print(f"ğŸš€ Starting asset move operation")
    print(f"   Source: {source_folder}")
    print(f"   Destination: {destination_folder}")
    print(f"   Delete source after copy: {delete_source}")
    print()
    
    ensure_folder_exists(destination_folder)
    move_recursive(source_folder, destination_folder)
    
    if delete_source:
        print("\nâœ… All assets moved successfully (copied and deleted from source).")
    else:
        print("\nâœ… All assets copied successfully (source preserved).")
```

## Enter source and destination paths, then run

Set `delete_source=True` for a true move operation (copy + delete source).

Set `delete_source=False` to only copy assets, preserving the source folder.

```python
# Define source and destination folders
source_folder = "projects/ee-okavango/assets/water_masks/monthly_DSWE_Landsat_30m_v3/Source_LS_Composites"
destination_folder = "projects/ee-okavango/assets/water_masks/monthly_DSWE_Landsat_30m_v4/Source_LS_Composites"

# Run the move operation
move_assets_recursively(source_folder, destination_folder, delete_source=True)
```
